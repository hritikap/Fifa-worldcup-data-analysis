### Also we can add a new column 'age' for furthermore analysis of the datasets. 
#### For this  we will use a dataframe having a column of date of birth and convert it to age and add that column to that dataframe.

```python
from datetime import datetime, date

def age(born):
    born = datetime.strptime(born, "%d/%m/%Y").date()
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
```

# Countries and the number of times they have won different awards of the worldcup represented in Bar Chart

#### At first lets count the total no of times the countries have won each awards in the world cup

```python
Country_Goldenboot=Golden_boot['National_team'].reset_index()
Country_Goldenboot['Golden_bootwon']=Country_Goldenboot.groupby(['National_team']).transform('count')
Country_Goldenboot=Country_Goldenboot.drop_duplicates(subset='National_team')
Country_Goldenboot=Country_Goldenboot.drop(['index'],axis=1).set_index('National_team')
Country_Goldenboot
```

```python
Country_Goldenball=Golden_ball['National_team'].reset_index()
Country_Goldenball['Golden_ballwon']=Country_Goldenball.groupby(['National_team']).transform('count')
Country_Goldenball=Country_Goldenball.drop_duplicates(subset='National_team')
Country_Goldenball=Country_Goldenball.drop(['index'],axis=1).set_index('National_team')
Country_Goldenball
```

```python
Country_goldenglove=Golden_glove['National_team'].reset_index()
Country_goldenglove['Golden_glovewon']=Country_goldenglove.groupby('National_team').transform('count')
Country_goldenglove=Country_goldenglove.drop_duplicates(subset='National_team')
Country_goldenglove=Country_goldenglove.drop(['index'],axis=1).set_index('National_team')
Country_goldenglove
```

```python
Country_youngestplayer=Fifa_youngplayer['National_team'].reset_index()
Country_youngestplayer['Young_awardwon']=Country_youngestplayer.groupby(['National_team']).transform('count')
Country_youngestplayer=Country_youngestplayer.drop_duplicates(subset='National_team')
Country_youngestplayer=Country_youngestplayer.drop(['index'],axis=1).set_index('National_team')
Country_youngestplayer
```

#### Now lets combine all the awards that the each countries have won into a single dataframe and plot a barchart according to that dataframe

```python
Combined_country_awardswinners=pd.concat([Country_Goldenboot,Country_Goldenball,Country_goldenglove,Country_youngestplayer],axis=1)
Combined_country_awardswinners=Combined_country_awardswinners.fillna(0).astype(int)
Combined_country_awardswinners = Combined_country_awardswinners.loc[:,~Combined_country_awardswinners.apply(lambda x: x.duplicated(),axis=1).all()].copy()
Combined_country_awardswinners
```

```python
import seaborn 
seaborn.set() 
Combined_country_awardswinners.plot(kind='barh',width=0.7,figsize=(10,25),lw=2,grid=False)
plt.title('Countries that have won awards in Fifa worldcups')
plt.xlim(0,5,10)
plt.xlabel('Number of times award won')
plt.ylabel('Awards winning countries')
```

#### The above barplot represents all those countries and the number time they have won the pretigious awards of the World cup ie 'The GoldenBoot','The GoldenBall','The GoldenGlove' and 'The Youngestplayer' award.

#### It can be observed that the most award winning countries are Italy, Germany, France and Argentina ie these four countries are the ones that have won all the four prestigious awards of the world cup.


# Pie Chart representation of Total awards winners of the Worldcup

```python
Combined_country_totalwinners=Combined_country_awardswinners.copy().reset_index()
Combined_country_totalwinners['Total_won']=Combined_country_totalwinners.sum(axis=1,numeric_only=True)
Combined_country_totalwinners
```

```python
labels=Combined_country_totalwinners['National_team']
sizes=Combined_country_totalwinners['Total_won']
fig1, ax1 = plt.subplots(figsize=(15, 15))
ax1.pie(sizes,labels=labels,autopct='%1.1f%%',
        shadow=True, startangle=90, textprops={'size': 'smaller','weight':'bold'})
ax1.axis('equal')
plt.title('Percentage of total awards won by countries in the  worldcup till now \n \n',fontsize=20)
plt.show()
```

#### The above pie chart represents the percentage of total awards that a country has won in the worldcup till date.
#### So from the above chart we can observe that the most award winning country is lead by Germany with total of 15.5% awards which is followed by Argentina and france as the second most award wining countries with the total of 12.1% awards

# Distribution of Players height(cm) using Histogram

```python
Playerheight=players_df.sort_values(by=['HEIGHT (CM)'])[['HEIGHT (CM)']].reset_index()
Playerheight=Playerheight.drop('index',axis=1)
Playerheight
```

```python
plt.hist(Playerheight,bins=8)
plt.title('Total no of players with different range of height',fontsize=20)
plt.xlim(160,205,10)
plt.ylabel('Frequency of the players')
plt.xlabel('height of the players (cm)')
```

#### The above histogram represents the height(in cm) of all the players in the world cup 2022. Here we have eight classes(165 to just below 170cm,170cm to  just below 175cm etc) each of which has an interval of 5cm. The height of each column represents the number of player that fall into each class.
#### So the highest number of players falls in the class (185cm to just below 190cm) and the lowest number of players falls in the class of(195cm to just below 205cm). That means most of the players height is between 185cm to 190cm.
